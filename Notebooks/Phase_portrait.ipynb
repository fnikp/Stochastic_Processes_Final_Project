```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1, X2 = np.meshgrid(np.linspace(-4, 4, 200), np.linspace(-4, 4, 200))


def dynamics(x1, x2):
    return [4 * x1 - x1 ** 3,
            -x1 * x2]

dx1, dx2 = dynamics(X1, X2)
norm = np.sqrt(dx1**2 + dx2**2)

fig, ax = plt.subplots(figsize=(12, 10))

cs = ax.contourf(X1, X2, norm, cmap ="viridis")
cbar = plt.colorbar(cs)

ax.streamplot(X1, X2, dx1, dx2, color='white', linewidth=1.3, density=1)

ax.axvline(x=-2, color='orange', ls='--')
ax.axvline(x=0, color='orange', ls='--')
ax.axvline(x=2, color='orange', ls='--')

ax.axhline(y=0, color='orange', ls='--')

plt.xlabel('$x_1$', fontsize=25)
plt.ylabel('$x_2$', fontsize=25)

plt.show()
```
